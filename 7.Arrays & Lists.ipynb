# Array Sum

Given an array of length N, you need to find and print the sum of all elements of the array.<br>
<br>
Input Format :<br>
Line 1 : An Integer N i.e. size of array<br>
Line 2 : N integers which are elements of the array, separated by spaces<br>
<br>
Output Format :<br>
Sum<br>
<br>
Constraints :<br>
1 <= N <= 10^6<br>
<br>
Sample Input :<br>
3<br>
9 8 9<br>
Sample Output :<br>
26

```python
p=input()
arr = input().split()
sum = 0
for x in range(len(arr)):
    p=int(arr[x])
    sum=sum+p
print(sum)
```

# Swap Alternate

You have been given an array/list(ARR) of size N. You need to swap every pair of alternate elements in the array/list.<br>
<br>
You don't need to print or return anything, just change in the input array itself.<br>
<br>
Input Format :<br>
The first line contains an Integer 't' which denotes the number of test cases or queries to be run. Then the test cases follow.<br>
<br>
First line of each test case or query contains an integer 'N' representing the size of the array/list.<br>
<br>
Second line contains 'N' single space separated integers representing the elements in the array/list.<br>
<br>
Output Format :<br>
For each test case, print the elements of the resulting array in a single row separated by a single space.<br>
<br>
Output for every test case will be printed in a separate line.<br>
<br>
Constraints :<br>
1 <= t <= 10^2<br>
0 <= N <= 10^5<br>
Time Limit: 1sec<br>
<br>
Sample Input 1:<br>
1<br>
6<br>
9 3 6 12 4 32<br>
Sample Output 1 :<br>
3 9 12 6 32 4<br>
Sample Input 2:<br>
2<br>
9<br>
9 3 6 12 4 32 5 11 19<br>
4<br>
1 2 3 4<br>
Sample Output 2 :<br>
3 9 12 6 32 4 11 5 19 <br>
2 1 4 3

```python
def swapAlternate (li):
    length =len(li)
    i = 0
    for k in range(length//2):
        li[i],li[i+1]=li[i+1],li[i]
        i+=2
    print(*li,end=" ")
    print()

loop_cycle=int(input())
for i in range(0, loop_cycle):
    faltu = input()
    lip = [int(x) for x in input().split()]
    swapAlternate(lip)
```

# Find Duplicate

You have been given an integer array/list(ARR) of size N which contains numbers from 0 to (N - 2). Each number is present at least once. That is, if N = 5, the array/list constitutes values ranging from 0 to 3 and among these, there is a single integer value that is present twice. You need to find and return that duplicate number present in the array.<br>
<br>
Note :<br>
Duplicate number is always present in the given array/list.<br>
<br>
Input format :<br>
The first line contains an Integer 't' which denotes the number of test cases or queries to be run. Then the test cases follow.<br>
<br>
First line of each test case or query contains an integer 'N' representing the size of the array/list.<br>
<br>
Second line contains 'N' single space separated integers representing the elements in the array/list.<br>
Output Format :<br>
For each test case, print the duplicate element in the array/list.<br>
<br>
Output for every test case will be printed in a separate line.<br>
<br>
Constraints :<br>
1 <= t <= 10^2<br>
0 <= N <= 10^3<br>
Time Limit: 1 sec<br>
<br>
Sample Input 1:<br>
1<br>
9<br>
0 7 2 5 4 7 1 3 6<br>
Sample Output 1:<br>
7<br>
Sample Input 2:<br>
2<br>
5<br>
0 2 1 3 1<br>
7<br>
0 3 1 5 4 3 2<br>
Sample Output 2:<br>
1<br>
3

```python
N = int(input())
k=0
while k<N:
    n = int(input())
    li = [int(x) for x in input().split()]
    new_li = li.copy()
    for i in range(n):
        new_li[i] = -1
        if li[i] in new_li:
            print(li[i])
            break
    k+=1
```

# Array Intersection

You have been given two integer arrays/list(ARR1 and ARR2) of size M and N, respectively. You need to print their intersection; An intersection for this problem can be defined when both the arrays/lists contain a particular value or to put it in other words, when there is a common value that exists in both the arrays/lists.<br>
<br>
Note :<br>
Input arrays/lists can contain duplicate elements.<br>
<br>
The intersection elements printed would be in the order they appear in the first array/list(ARR1)<br>
<br>
<br>
Input format :<br>
The first line contains an Integer 't' which denotes the number of test cases or queries to be run. Then the test cases follow.<br>
<br>
First line of each test case or query contains an integer 'N' representing the size of the first array/list.<br>
<br>
Second line contains 'N' single space separated integers representing the elements of the first the array/list.<br>
<br>
Third line contains an integer 'M' representing the size of the second array/list.<br>
<br>
Fourth line contains 'M' single space separated integers representing the elements of the second array/list.<br>
<br>
Output format :<br>
For each test case, print the intersection elements in a row, separated by a single space.<br>
<br>
Output for every test case will be printed in a separate line.<br>
<br>
Constraints :<br>
1 <= t <= 10^2<br>
0 <= N <= 10^5<br>
0 <= M <= 10^5<br>
Time Limit: 1 sec <br>
<br>
Sample Input 1 :<br>
2<br>
6<br>
2 6 8 5 4 3<br>
4<br>
2 3 4 7 <br>
2<br>
10 10<br>
1<br>
10<br>
Sample Output 1 :<br>
2 4 3<br>
10<br>
Sample Input 2 :<br>
1<br>
4<br>
2 6 1 2<br>
5<br>
1 2 3 4 2<br>
<br>
Sample Output 2 :<br>
2 1 2<br>
Explanation for Sample Output 2 :<br>
Since, both input arrays have two '2's, the intersection of the arrays also have two '2's. The first '2' of first array matches with the first '2' of the second array. Similarly, the second '2' of the first array matches with the second '2' if the second array.

<b>Note there is some Problem in running this code in jupyter notebook Run in Pycharm it will fine
<br>or you copy paste directly in CodingNinjas</b>

```python
import sys

def intersections(arr1, n, arr2, m):
    for i in range(n):
        for j in range(m):
            if arr1[i] == arr2[j]:
                print(arr1[i],end =" ")
                arr2[j] = sys.maxsize
                break
               

def takeInput() :
    n = int(sys.stdin.readline().strip())

    if n == 0 :
        return list(), 0

    arr = list(map(int, sys.stdin.readline().strip().split(" ")))
    return arr, n



#main
t = int(sys.stdin.readline().strip())

while t > 0 :
    arr1 , n = takeInput()
    arr2 , m = takeInput()
    intersections(arr1, n, arr2, m)
    print()
    
    t -= 1
```

# Pair Sum

You have been given an integer array/list(ARR) and a number X. Find and return the total number of pairs in the array/list which sum to X.<br>
<br>
Note:<br>
Given array/list can contain duplicate elements. <br>
<br>
Input format :<br>
The first line contains an Integer 't' which denotes the number of test cases or queries to be run. Then the test cases follow.<br>
<br>
First line of each test case or query contains an integer 'N' representing the size of the first array/list.<br>
<br>
Second line contains 'N' single space separated integers representing the elements in the array/list.<br>
<br>
Third line contains an integer 'X'.<br>
<br>
Output format :<br>
For each test case, print the total number of pairs present in the array/list.<br>
<br>
Output for every test case will be printed in a separate line.<br>
<br>
Constraints :<br>
1 <= t <= 10^2<br>
0 <= N <= 10^3<br>
0 <= X <= 10^9<br>
Time Limit: 1 sec<br>
<br>
Sample Input 1:<br>
1<br>
9<br>
1 3 6 2 5 4 3 2 4<br>
7<br>
Sample Output 1:<br>
7<br>
Sample Input 2:<br>
2<br>
9<br>
1 3 6 2 5 4 3 2 4<br>
12<br>
6<br>
2 8 10 5 -2 5<br>
10<br>
Sample Output 2:<br>
0<br>
2<br>
Explanation for Input 2:<br>
Since there doesn't exist any pair with sum equal to 12 for the first query, we print 0.<br>
For the second query, we have 2 pairs in total that sum up to 10. They are, (2, 8) and (5, 5).

```python
def duplicate (list1,sum):
    sum_time=0
    size=len(list1)
    for i in range(0,size):
        for j in range(i+1,size):
            if list1[i] + list1[j]==sum:
                sum_time+=1
            else:
                continue
    print(sum_time)

n=int(input()) #First Input
for i in range(n):
    faltu1=int(input()) #Secound input #no of ele in list
    if faltu1!=0:
        list = [int(x) for x in input().split()]
    X=int(input()) #3rd Input
    if faltu1!=0 :
        duplicate(list,X)
    else:
        print(0)
```

# Triplet Sum

You have been given a random integer array/list(ARR) and a number X. Find and return the triplet(s) in the array/list which sum to X.<br>
<br>
Note :<br>
Given array/list can contain duplicate elements.<br>
<br>
Input format :<br>
The first line contains an Integer 't' which denotes the number of test cases or queries to be run. Then the test cases follow.<br>
<br>
First line of each test case or query contains an integer 'N' representing the size of the first array/list.<br>
<br>
Second line contains 'N' single space separated integers representing the elements in the array/list.<br>
<br>
Third line contains an integer 'X'.<br>
<br>
Output format :<br>
For each test case, print the total number of triplets present in the array/list.<br>
<br>
Output for every test case will be printed in a separate line.<br>
<br>
Constraints :<br>
1 <= t <= 50<br>
0 <= N <= 10^2<br>
0 <= X <= 10^9<br>
Time Limit: 1 sec<br>
<br>
Sample Input 1:<br>
1<br>
7<br>
1 2 3 4 5 6 7 <br>
12<br>
Sample Output 1:<br>
5<br>
Sample Input 2:<br>
2<br>
7<br>
1 2 3 4 5 6 7 <br>
19<br>
9<br>
2 -5 8 -6 0 5 10 11 -3<br>
10<br>
Sample Output 2:<br>
0<br>
5<br>
Explanation for Input 2:<br>
Since there doesn't exist any triplet with sum equal to 19 for the first query, we print 0.<br>
<br>
For the second query, we have 5 triplets in total that sum up to 10. They are, (2, 8, 0), (2, 11, -3), (-5, 5, 10), (8, 5, -3) and (-6, 5, 11)

```python
def duplicate (list1,sum):
    sum_time=0
    n=len(list1)
    for i in range(0,n):
        for j in range(i+1,n):
            for k in range(j+1,n):
                if list1[i] + list1[j]+ list1[k]==sum:
                    sum_time+=1
                else:
                    continue
    print(sum_time)

n=int(input())
for i in range(n):
    faltu1=int(input())
    if faltu1!=0:
        list = [int(x) for x in input().split()]
    X=int(input())
    if faltu1!=0 :
        duplicate(list,X)
    else:
        print(0)
```
